#Trabajo con los datos

```python
# Voy a leer la tabla con los estados desde una URL
url = 'https://es.wikipedia.org/wiki/Estado_de_los_Estados_Unidos'
df_estados = pd.read_html(url)
df_estados = df_estados[1] #Obtengo la tabla donde estan todos los estados que necesito
lista_estados = df_estados['Estado'].to_numpy().transpose().tolist()
```

```python
#Agrego una nueva columna en el dataFrame principal usando Random Choice
#Si tiene un contrato hibrido se deja el valor siempre para Texas, en caso de remoto toma un valor de la lista de estados aleatoriamente
estado = 'Texas'
df_principal['States'] = df_principal['type'].apply(lambda x: estado if x == 'HYBRID' else np.random.choice(lista_estados))

# Limpio caracteres especiales y números de la columna 'States'
df_principal['States'] = df_principal['States'].str.replace(r'\[.*?\]', '', regex=True).str.replace(r'\d+', '', regex=True)
```
